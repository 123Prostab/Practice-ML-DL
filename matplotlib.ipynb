```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("indices")
plt.title("Myplot")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel("squares")
plt.xlabel("numbers")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2)
plt.plot(t,t**2,"b--",label='^2')
plt.plot(t,t**2.2,"rs",label='^2.2')
plt.plot(t,t**2.5,"g^",label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth = 5.0)
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color = "r",linewidth = 2.0)
plt.setp(lines[1],color = "g",linewidth = 2.0)
plt.grid()
plt.show()
```
